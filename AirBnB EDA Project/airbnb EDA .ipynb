# EDA

  ***TASK5: DATA ANALYSIS***

```python
data = df[df['price']<1500]
sns.boxplot(data = data , x ='price')
```

```python
plt.figure(figsize =(8,9))
plt.ylabel("Frequency")
plt.title("Price Distribution")
sns.histplot(data =data, x='price', bins = 100)
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize =(6,3))
plt.ylabel("Frequency")
plt.title("availability_365 Distribution")
sns.histplot(data =data, x='availability_365')
plt.show()
```

```python
df.groupby(by = 'neighbourhood_group') ['price'].mean()
```

```python
#average price per bed 
df.groupby(by = 'neighbourhood_group') ['price per bed'].mean()
```

```python
df['price per bed'] = df['price']/ df['beds']
df.head()
```

```python
# price dependancy on neighbourhood group

plt.figure(figsize =(6,5))
sns.barplot(data =df , x= 'neighbourhood_group' ,y ='price', hue= 'room_type')
plt.show()
```

```python
# number of reviews and price relationship 
plt.figure(figsize =(6,5))
plt.title("Locality and Review Analysis")
sns.scatterplot(data = data , x= 'number_of_reviews', y ='price', hue ='neighbourhood_group')
plt.show()
```

```python
sns.pairplot(data = data ,vars=['price','minimum_nights', 'number_of_reviews', 'availability_365'], hue= 'room_type')
```

```python
# geographical distribution of airbnb listing
sns.scatterplot(data = data , x='longitude',y ='latitude', hue  ='room_type')
```

```python
# heat map - correlation of one variable with another numeric columns
corr = data[['latitude', 'longitude', 'price','minimum_nights', 'number_of_reviews', 'reviews_per_month','availability_365','beds']].corr()
corr
sns.heatmap(data = corr , annot= True)
plt.figure(figsize =(8,7))
```
